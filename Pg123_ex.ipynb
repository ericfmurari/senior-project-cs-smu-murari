```python
m=320
k=16000
B=1200
```

```python
Lambda=-B/(2*m)+n.array([-1,1])*sqrt(B^2-4*m*k)/(2*m);
sigma=n.real(Lambda[0])
omega=n.imag(Lambda[1])
K=m/k
C3=-m/k
C4=sigma/omega*m/k
A1=sqrt(C3*C3+C4*C4)
psi1=-n.arctan2(C4,C3)
A2=sqrt(sigma*sigma+omega*omega)
psi2=-n.arctan2(-omega,sigma)
print(psi2)
```

```python
vm= lambda t: A1*A2*B/m*n.exp(sigma*t)*n.cos(omega*t+psi1+psi2)+A1*k/m*n.exp(sigma*t)*n.cos(omega*t+psi1)+1
```

```python
t_a=n.linspace(0,3,200)
plt.plot(t_a,vm(t_a))
plt.xlabel("time (s)")
plt.ylabel('velocity $v_m(t)$ (m/s)')
plt.grid()
```
